# Import packages

```python
import pandas
import os
import matplotlib.pyplot as plt
```

# Load dataset

```python
result = pandas.read_csv('july.csv')
result
```

# Fix the data where quotation are missed

```python
result['item_name_en']=result['item_name_en'].apply(lambda x: x.replace('"', ''))
result['item_name_en'] = result['item_name_en'].apply(lambda x: '"' + str(x) + '"')
result
```

# Remove unwanted columns

```python
result = result.drop(["invoice_detail_item_no", "invoice_date"], axis=1)
result
```

# Join same invoice items

```python
result=result.groupby('invoice_detail_invoice_no')['item_name_en'].apply(' '.join).reset_index()
result
result = result.drop(["invoice_detail_invoice_no"], axis=1)

result
```

```python
result.to_csv("transactionProcessed.csv", index=False,header=False)
```
