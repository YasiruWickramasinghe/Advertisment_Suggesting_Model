# Import packages

```python
import pandas
import os
import matplotlib.pyplot as plt
```

# Load dataset

```python
result = pandas.read_csv('january.csv')
result
```

# Fix the data where quotation are missed

```python
result['item_name_en']=result['item_name_en'].apply(lambda x: x.replace('"', ''))
result['item_name_en'] = result['item_name_en'].apply(lambda x: '"' + str(x) + '"')
result
```

# Remove unwanted columns

```python
result = result.drop(["invoice_detail_item_no", "invoice_date"], axis=1)
result
```

# Join same invoice items

```python
result=result.groupby('invoice_detail_invoice_no')['item_name_en'].apply(' '.join).reset_index()
result
result = result.drop(["invoice_detail_invoice_no"], axis=1)

result
```

```python
result.to_csv("preprocessed_jan.csv", index=False,header=False)
```

# Automate the process

```python
import pandas
import os
import matplotlib.pyplot as plt

fileList = []
for path, folders, files in os.walk('./dataset'):
        for file in files:
            fileList.append(os.path.join(path, file))
for filename in fileList:
    path=str(filename).strip('[]').split("./dataset/")[1]
    print(path)
    result = pandas.read_csv(filename)
    result['item_name_en']=result['item_name_en'].apply(lambda x: x.replace('"', ''))
    result['item_name_en'] = result['item_name_en'].apply(lambda x: '"' + str(x) + '"')
    result = result.drop(["invoice_detail_item_no", "invoice_date"], axis=1)
    result=result.groupby('invoice_detail_invoice_no')['item_name_en'].apply(' '.join).reset_index()
    result = result.drop(["invoice_detail_invoice_no"], axis=1)
    result.to_csv('./dataset/preprocesesd_data/preprocessed_'+path,index=False,header=False)
```

# Dataset Stats

```python
fileList = []
month_records={}
for path, folders, files in os.walk('./dataset/preprocesesd_data'):
        for file in files:
            fileList.append(os.path.join(path, file))
for filename in fileList:
    month=str(filename).strip('[]').split("./dataset/preprocesesd_data/preprocessed_")[1].split(".csv")[0]
    result = pandas.read_csv(filename)
    print("Number of records for month "+month+" : ", len(result.index))
    month_records[month] = len(result.index)
    
month_ordered=['january','february','march','april','may','june','july','august','september','octomber','november','december']
records_ordered=[]

for month in month_ordered:
    for key, value in month_records.items():
        if(month==key):
            records_ordered.append(value)
    
    


    
```

```python
fig = plt.figure(figsize=(9, 6))
ax = fig.add_axes([0,0,1,1])
ax.bar(month_ordered,records_ordered)
plt.setp(ax.get_xticklabels(), rotation=30, horizontalalignment='right')
plt.title('Dataset distribution against months',fontweight='bold')
plt.xlabel("month")
plt.ylabel("number of records")
plt.savefig('./graphs/data_distribution.jpg', dpi=300, bbox_inches = 'tight')
```
